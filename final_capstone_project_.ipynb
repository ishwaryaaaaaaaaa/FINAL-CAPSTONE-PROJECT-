```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("/content/dataset.csv")

# === Subtask 1: Preprocessing ===

# Combine date and time into a Timestamp
df["Timestamp"] = pd.to_datetime(df["LastUpdatedDate"] + " " + df["LastUpdatedTime"], format="%d-%m-%Y %H:%M:%S")

# Calculate Occupancy Rate
df["OccupancyRate"] = df["Occupancy"] / df["Capacity"]

# Map traffic level
traffic_map = {"low": 1, "medium": 2, "high": 3}
df["TrafficLevel"] = df["TrafficConditionNearby"].map(traffic_map)

# Assign vehicle type weight
vehicle_weights = {"bike": 0.5, "car": 1.0, "truck": 1.5}
df["VehicleTypeWeight"] = df["VehicleType"].map(vehicle_weights)

# Ensure IsSpecialDay is integer
df["IsSpecialDay"] = df["IsSpecialDay"].astype(int)

# === Subtask 2: Baseline Linear Pricing ===

# Sort for time-based simulation
df = df.sort_values(by=["SystemCodeNumber", "Timestamp"]).copy()

# Initialize Price column
df["Price"] = np.nan
initial_price = 10.0
alpha = 2.0  # You can tune this

# Apply pricing model per parking lot
for lot_id, group in df.groupby("SystemCodeNumber"):
    group = group.sort_values("Timestamp").copy()
    prices = [initial_price]
    for i in range(1, len(group)):
        prev_price = prices[-1]
        occ_rate = group.iloc[i - 1]["OccupancyRate"]
        new_price = prev_price + alpha * occ_rate
        new_price = np.clip(new_price, 5, 20)
        prices.append(new_price)
    df.loc[group.index, "Price"] = prices

# Preview results
df[["SystemCodeNumber", "Timestamp", "OccupancyRate", "Price"]].head(10)
```

```python
def normalize_demand(demands):
    min_d = min(demands)
    max_d = max(demands)
    return [(d - min_d) / (max_d - min_d) if max_d != min_d else 0 for d in demands]
```

```python
def compute_demand(occupancy, capacity, queue, traffic, is_special_day, vehicle_type_weight,
                   alpha, beta, gamma, delta, epsilon):
    utilization = occupancy / capacity
    demand = (alpha * utilization +
              beta * queue -
              gamma * traffic +
              delta * is_special_day +
              epsilon * vehicle_type_weight)
    return demand
```

```python
# === Subtask 3: Demand-Based Pricing Model ===

# Copy to preserve previous results
df_demand = df.copy()

# Step 1: Calculate Demand Score
alpha = 2.0
beta = 0.5
gamma = 1.0
delta = 1.5
epsilon = 1.0

df_demand["DemandRaw"] = (
    alpha * df_demand["OccupancyRate"] +
    beta * df_demand["QueueLength"] -
    gamma * df_demand["TrafficLevel"] +
    delta * df_demand["IsSpecialDay"] +
    epsilon * df_demand["VehicleTypeWeight"]
)

# Step 2: Normalize Demand (Min-Max Scaling per lot)
df_demand["DemandNorm"] = df_demand.groupby("SystemCodeNumber")["DemandRaw"].transform(
    lambda x: (x - x.min()) / (x.max() - x.min() + 1e-6)
)

# Step 3: Calculate Final Price
lambda_ = 0.8
df_demand["Price_DemandModel"] = 10 * (1 + lambda_ * df_demand["DemandNorm"])

# Step 4: Clip Prices between $5 and $20
df_demand["Price_DemandModel"] = df_demand["Price_DemandModel"].clip(lower=5, upper=20)

# Preview output
df_demand[["SystemCodeNumber", "Timestamp", "DemandRaw", "DemandNorm", "Price_DemandModel"]].head(10)
```

```python
#model - 3
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c
```

```python
# Get unique parking lots and their lat-long
lot_info = df_demand.groupby("SystemCodeNumber")[["Latitude", "Longitude"]].first()

# Find nearby lots within 1 km
distance_threshold = 1.0
nearby_lots = {}

for lot1 in lot_info.index:
    nearby = []
    lat1, lon1 = lot_info.loc[lot1]
    for lot2 in lot_info.index:
        if lot1 == lot2:
            continue
        lat2, lon2 = lot_info.loc[lot2]
        dist = haversine(lat1, lon1, lat2, lon2)
        if dist <= distance_threshold:
            nearby.append(lot2)
    nearby_lots[lot1] = nearby
```

```python
df_competitive = df_demand.copy()
df_competitive = df_competitive.sort_values(["SystemCodeNumber", "Timestamp"]).copy()

adjusted_prices = []

for idx, row in df_competitive.iterrows():
    lot = row["SystemCodeNumber"]
    timestamp = row["Timestamp"]
    occ_rate = row["OccupancyRate"]
    price = row["Price_DemandModel"]

    # Nearby prices at the same timestamp
    competitors = nearby_lots.get(lot, [])
    nearby_prices = df_competitive[
        (df_competitive["SystemCodeNumber"].isin(competitors)) &
        (df_competitive["Timestamp"] == timestamp)
    ]["Price_DemandModel"]

    if not nearby_prices.empty:
        min_nearby = nearby_prices.min()
        max_nearby = nearby_prices.max()

        # If full and others are cheaper → drop slightly
        if occ_rate >= 0.9 and price > min_nearby:
            price -= 1.0

        # If room and others are expensive → increase slightly
        elif occ_rate <= 0.5 and price < max_nearby:
            price += 1.0

    # Keep price within bounds
    price = np.clip(price, 5, 20)
    adjusted_prices.append(price)

df_competitive["Price_CompetitiveModel"] = adjusted_prices
```

```python
import pathway as pw
import pandas as pd
import numpy as np

# === Step 1: Define a Schema ===
class ParkingSchema(pw.Schema):
    SystemCodeNumber: str
    Capacity: int
    Latitude: float
    Longitude: float
    Occupancy: int
    VehicleType: str
    TrafficConditionNearby: str
    QueueLength: int
    IsSpecialDay: int
    LastUpdatedDate: str
    LastUpdatedTime: str

# === Step 2: Read the CSV in real-time mode ===
parking_data = pw.io.csv.read(
    "./dataset.csv",  # Or your path
    schema=ParkingSchema,
    mode="streaming",
    autocommit_duration_ms=500  # Simulates streaming every 500ms
)

# === Step 3: Combine and Format Timestamp ===
@pw.udf
def combine_timestamp(date, time):
    return pd.to_datetime(date + " " + time, format="%d-%m-%Y %H:%M:%S")

parking_data = parking_data.with_columns(
    Timestamp=combine_timestamp(parking_data.LastUpdatedDate, parking_data.LastUpdatedTime)
)

# === Step 4: Define your Demand-Based Pricing Logic ===
@pw.udf
def compute_demand_price(occ, cap, queue, traffic, special, vtype):
    alpha, beta, gamma, delta, epsilon = 2.0, 0.5, 1.0, 1.5, 1.0
    lambda_ = 0.8

    occ_rate = occ / cap if cap != 0 else 0
    vtype_weight = {"bike": 0.5, "car": 1.0, "truck": 1.5}.get(vtype, 1.0)
    traffic_level = {"low": 1, "medium": 2, "high": 3}.get(traffic, 2)

    demand = (
        alpha * occ_rate +
        beta * queue -
        gamma * traffic_level +
        delta * special +
        epsilon * vtype_weight
    )

    # Normalize demand with bounding instead of full scaling
    norm_demand = max(0, min(1, (demand - 1) / 10))
    price = 10 * (1 + lambda_ * norm_demand)
    return float(np.clip(price, 5, 20))

# === Step 5: Apply the UDF to generate pricing ===
output_table = parking_data.select(
    lot=parking_data.SystemCodeNumber,
    timestamp=parking_data.Timestamp,
    price=compute_demand_price(
        parking_data.Occupancy,
        parking_data.Capacity,
        parking_data.QueueLength,
        parking_data.TrafficConditionNearby,
        parking_data.IsSpecialDay,
        parking_data.VehicleType
    )
)

# === Step 6: Print result to console in real time ===
pw.io.jsonlines.write(output_table,"streamed_prices.jsonl")

# === Step 7: Run the Pipeline ===
pw.run()
```
